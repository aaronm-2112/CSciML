```python
# Common imports
import numpy as np
import os
import pandas as pd

%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12

print("Basic libraries imported successfully")
```

```python
#basic numpy array manipulation & properties
array1 = np.arange(10).reshape(2,5)
print("Array1 shape is: " + str(array1.shape))
print("Array1's dimensions: " + str(array1.ndim))
print("contents of array1 are: " + str(array1))

print(" ")

print("Can create a python list from a numpy array and vice versa")
print("Array1's type: " + str(type(array1)))
list1 = array1.tolist()
print("list1 type: " + str(type(list1)))

print(" ")

print("Existing arrays can be reshaped" )
array2 = array1.reshape(1,10)
print("1 dimensional shape of array1 in array2: " + str(array2))

print(" ")

#Basic operations on numpy arrays
array1 *= 3
print("Array1 after elementwise multiplication by 3: " + str(array1))
array3 = np.arange(10).reshape(2,5)
array1 += array3
print("Array1 after elementwise addition with another array: " + str(array1))

print(" ")

print("operations on specific np array axes")
print("Summing array1's 0 axis together: " + str(array1.sum(axis = 0)))
print("Summing array1's 1 axis together: " + str(array1.sum(axis = 1)))
print("Finding the min of each column of an array: " + str(array1.min(axis = 0)))
print("Finding the min of each row of an array: " + str(array1.min(axis = 1)))
```

```python
#basic matplotlib functions 
plt.plot([1,2,3,4], [1,2,4,8]) #x-axis on left y-axis on right
```

```python
#basic matplotlib functions 
#plotting with different markers and multiple plots on one graph using numpy arrays for axis values

plotArray1 = np.arange(1.,8.,0.3) #start, stop, step 
plt.plot(plotArray1, plotArray1, "r--", plotArray1, plotArray1**2, "bs", plotArray1, plotArray1**3, "g^")
plt.ylabel("nonsense y value")
plt.xlabel("nonsense x value")
```

```python
#basic myplotlib
#mltiple plots per figure is possible
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(1, (15,5)) #width of figure, height of figure, in inches

plt.subplot(2,3,1) #rows, columns, starting position in the figure (left to right)
plt.bar(names, values)
plt.subplot(2,3,2)
plt.scatter(names, values)
plt.subplot(2,3,3)
plt.plot(names, values)
plt.subplot(2,3,4)
plt.plot(["bird", "donkey", "cow"], [1, 200, 350])
plt.xlabel("animals")
plt.ylabel("# of lbs")
plt.suptitle('Categorical Plotting')
plt.show()
```
